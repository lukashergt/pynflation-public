# Fraction of KD region on Planck surface

```python
mass_scales = np.concatenate(( np.logspace(-2, np.log10(1 / np.sqrt(51)), 50), np.logspace(np.log10(1 / np.sqrt(51)), 0, 50), np.logspace(0, 1, 50 + 1) ))
```

```python
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(8, 3), gridspec_kw=dict(width_ratios=[1.0, 1.962, 1.962]), sharey='row')
fig.subplots_adjust(hspace=0, wspace=0)

ax[0, 0].set_title("Quadratic"                         , color=cKD_m2phi2)
ax[0, 1].set_title("Plateau"                           , color=cKD_starob)
ax[0, 2].set_title("Double-Well,  $\mu=5~\mathrm{m_p}$", color=cKD_dbwell)

ax[0, 2].text(x=1.05, y=0.5, s="full $\phi$-range"     , va='center', ha='center', rotation='vertical', transform=ax[0, 2].transAxes)
ax[1, 2].text(x=1.05, y=0.5, s="effective $\phi$-range", va='center', ha='center', rotation='vertical', transform=ax[1, 2].transAxes)

ax[0, 0].set_xlim(5e-3, 2e1)
ax[0, 0].loglog(mass_scales, KDFraction(['m2phi2', mass_scales]).frac_uni(), ls='-' , c=mpl.cm.Blues(0.4) )#, label="uniform on $\phi$")
ax[0, 0].loglog(mass_scales, KDFraction(['m2phi2', mass_scales]).frac_log(), ls='--', c=mpl.cm.Blues(0.3) )#, label="logarithmic on $\phi$")
ax[0, 0].loglog(mass_scales, KDFraction(['m2phi2', mass_scales]).frac_eng(), ls='-.', c=mpl.cm.Blues(0.7) )#, label="uniform in energy")
ax[0, 0].loglog(mass_scales, KDFraction(['m2phi2', mass_scales]).frac_arc(), ls=':' , c=mpl.cm.Blues(0.99))#, label="arclength")
ax[0, 0].set_yticks([0, 0.25, 0.5, 0.75, 1.0])
ax[0, 0].set_xticks([1e-2, 1e-1, 1e0, 1e1])
ax[0, 0].set_ylabel("$f_\mathrm{KD}$")
ax[0, 0].set_xlabel("$m~/~\mathrm{m_p}\qquad$")
plt.setp(ax[0, 0].get_xticklabels()[-1], visible=False)

ax[0, 1].set_xlim(5e-3, 2e1)
ax[0, 1].loglog(mass_scales, KDFraction(['starobinsky', mass_scales]).frac_uni(), ls='-' , c=mpl.cm.Oranges(0.4))#, label="uniform on $\phi$")
ax[0, 1].loglog(mass_scales, KDFraction(['starobinsky', mass_scales]).frac_log(), ls='--', c=mpl.cm.Oranges(0.2))#, label="logarithmic on $\phi$")
ax[0, 1].loglog(mass_scales, KDFraction(['starobinsky', mass_scales]).frac_eng(), ls='-.', c=mpl.cm.Oranges(0.7))#, label="uniform in energy")
ax[0, 1].loglog(mass_scales, KDFraction(['starobinsky', mass_scales]).frac_arc(), ls=':' , c=mpl.cm.Oranges(0.9))#, label="arclength")
ax[0, 1].loglog(1, 1, ls='-' , c='0.6', label="uniform in $\phi$")
ax[0, 1].loglog(1, 1, ls='--', c='0.8', label="logarithmic in $\phi$")
ax[0, 1].loglog(1, 1, ls='-.', c='0.3', label="uniform in energy")
ax[0, 1].loglog(1, 1, ls=':' , c='0.1', label="uniform on Planck surface")
ax[0, 1].legend(ncol=4, loc='lower center', bbox_to_anchor=(0.75, 1.18))

ax[0, 2].set_xlim(5e-3, 2e1)
ax[0, 2].loglog(mass_scales, KDFraction(['hilltop', mass_scales, 5]).frac_uni(), ls='-' , c=mpl.cm.Greens(0.5) )#, label="uniform on $\phi$")
ax[0, 2].loglog(mass_scales, KDFraction(['hilltop', mass_scales, 5]).frac_log(), ls='--', c=mpl.cm.Greens(0.3) )#, label="logarithmic on $\phi$")
ax[0, 2].loglog(mass_scales, KDFraction(['hilltop', mass_scales, 5]).frac_eng(), ls='-.', c=mpl.cm.Greens(0.7) )#, label="uniform in energy")
ax[0, 2].loglog(mass_scales, KDFraction(['hilltop', mass_scales, 5]).frac_arc(), ls=':' , c=mpl.cm.Greens(0.99))#, label="uniform on surface")

fig.delaxes(ax[1, 0])

ax[1, 1].set_xlim(5e-3, 2e1)
ax[1, 1].loglog(mass_scales, KDFraction(['starobinsky', mass_scales]).frac_uni(use_eff_range=True), ls='-' , c=mpl.cm.Oranges(0.4))
ax[1, 1].loglog(mass_scales, KDFraction(['starobinsky', mass_scales]).frac_log(use_eff_range=True), ls='--', c=mpl.cm.Oranges(0.2))
ax[1, 1].loglog(mass_scales, KDFraction(['starobinsky', mass_scales]).frac_eng(use_eff_range=True), ls='-.', c=mpl.cm.Oranges(0.7))
ax[1, 1].loglog(mass_scales, KDFraction(['starobinsky', mass_scales]).frac_arc(use_eff_range=True), ls=':' , c=mpl.cm.Oranges(0.9))
ax[1, 1].tick_params('y', labelleft=True)
ax[1, 1].set_yticks([0, 0.25, 0.5, 0.75, 1])
ax[1, 1].set_ylabel("$f_\mathrm{KD}^\mathrm{eff}\qquad$")
ax[1, 1].set_xlabel("$\Lambda^2~/~\mathrm{m_p^2}$")
ax[1, 1].legend(title="$\phi>0$", loc='lower left', framealpha=0.5)
plt.setp(ax[1, 1].get_yticklabels()[-1], visible=False)

ax[1, 2].set_xlim(5e-3, 2e1)
ax[1, 2].loglog(mass_scales, KDFraction(['hilltop', mass_scales, 5]).frac_uni(use_eff_range=True), ls='-' , c=mpl.cm.Greens(0.5) )#, label="uniform on $\phi$")
ax[1, 2].loglog(mass_scales, KDFraction(['hilltop', mass_scales, 5]).frac_log(use_eff_range=True), ls='--', c=mpl.cm.Greens(0.3) )#, label="logarithmic on $\phi$")
ax[1, 2].loglog(mass_scales, KDFraction(['hilltop', mass_scales, 5]).frac_eng(use_eff_range=True), ls='-.', c=mpl.cm.Greens(0.7) )#, label="uniform in energy")
ax[1, 2].loglog(mass_scales, KDFraction(['hilltop', mass_scales, 5]).frac_arc(use_eff_range=True), ls=':' , c=mpl.cm.Greens(0.99))#, label="uniform on surface")
ax[1, 2].set_xlabel("$\Lambda^2~/~\mathrm{m_p^2}$")
ax[1, 2].legend(title="$|\phi|<\mu$", loc='lower left', framealpha=0.5)

for row in range(2):
    for col in range(3):
        ax[row, col].set_ylim(5e-3, 2e0)
        ax[row, col].axhspan(1, ax[0, 1].get_ylim()[1], color='0.5', alpha=0.3)
        ax[row, col].yaxis.set_major_formatter(mpl.ticker.FuncFormatter(lambda y, _: '{:g}'.format(y)))
        if row < 1 and col>=1:
            ax[row, col].set_xlabel("")
            ax[row, col].tick_params('x', labelbottom=False)
        if col >= 1 and row==0:
            ax[row, col].set_ylabel("")
            ax[row, col].tick_params('y', labelleft=False)

fig.savefig("figures/phasespace_KD_fractions.pdf", bbox_inches='tight')
fig.savefig("figures/phasespace_KD_fractions.png", bbox_inches='tight')
```

```python
fig = plt.figure(figsize=(8, 4))
ax = plt.gca()

ax.loglog(mass_scales, KDFraction(['m2phi2', mass_scales]).frac_uni(), lw=1.5, ls='-' , c=mpl.cm.Blues(0.4) , label="Quadratic, uniform in $\phi$")
ax.loglog(mass_scales, KDFraction(['m2phi2', mass_scales]).frac_log(), lw=1.5, ls='--', c=mpl.cm.Blues(0.3) , label="Quadratic, logarithmic in $\phi$")
ax.loglog(mass_scales, KDFraction(['m2phi2', mass_scales]).frac_eng(), lw=1.5, ls='-.', c=mpl.cm.Blues(0.7) , label="Quadratic, uniform in energy")
ax.loglog(mass_scales, KDFraction(['m2phi2', mass_scales]).frac_arc(), lw=1.5, ls=':' , c=mpl.cm.Blues(0.99), label="Quadratic, Planck surface")

ax.loglog(mass_scales, KDFraction(['starobinsky', mass_scales]).frac_uni(), lw=1.5, ls='-' , c=mpl.cm.Oranges(0.4), label="Plateau, uniform in $\phi$")
ax.loglog(mass_scales, KDFraction(['starobinsky', mass_scales]).frac_log(), lw=1.5, ls='--', c=mpl.cm.Oranges(0.2), label="Plateau, logarithmic in $\phi$")
ax.loglog(mass_scales, KDFraction(['starobinsky', mass_scales]).frac_eng(), lw=1.5, ls='-.', c=mpl.cm.Oranges(0.7), label="Plateau, uniform in energy")
ax.loglog(mass_scales, KDFraction(['starobinsky', mass_scales]).frac_arc(), lw=1.5, ls=':' , c=mpl.cm.Oranges(0.9), label="Plateau, Planck surface")

ax.loglog(mass_scales, KDFraction(['hilltop', mass_scales, 5]).frac_uni(), lw=1.5, ls='-' , c=mpl.cm.Greens(0.5) , label="Double-Well, uniform in $\phi$")
ax.loglog(mass_scales, KDFraction(['hilltop', mass_scales, 5]).frac_log(), lw=1.5, ls='--', c=mpl.cm.Greens(0.3) , label="Double-Well, logarithmic in $\phi$")
ax.loglog(mass_scales, KDFraction(['hilltop', mass_scales, 5]).frac_eng(), lw=1.5, ls='-.', c=mpl.cm.Greens(0.7) , label="Double-Well, uniform in energy")
ax.loglog(mass_scales, KDFraction(['hilltop', mass_scales, 5]).frac_arc(), lw=1.5, ls=':' , c=mpl.cm.Greens(0.99), label="Double-Well, Planck surface")

ax.set_ylim(5e-3, 1.25e0)
ax.set_xlabel("$m~/~\mathrm{m_p}\qquad$  or  $\qquad\Lambda^2~/~\mathrm{m_p^2}$")
ax.set_ylabel("$f_\mathrm{KD}$")

ax.axhspan(1, ax.get_ylim()[1], color='0.5', alpha=0.3)
ax.legend(ncol=3, loc='lower center', bbox_to_anchor=(0.5, 1));

fig.savefig("figures/phasespace_KD_fractions_in_one.pdf")
fig.savefig("figures/phasespace_KD_fractions_in_one.png")
```
